```python
import pandas as pd
```

```python
pecos = pd.read_json('data/parsed_output.json')
pecos.head()
```

```python
ontology_mentions = []
for index, example in pecos.iterrows():
    example_d = {
        'text': example['peco_statement'],
        'url': example['url'],
    }
    elements = example['peco_elements']
    for element, elementdata in elements.items():
        for d in elementdata:
            matches = d['matches']
            for match in matches:
                ontology = match['ontology_name']
                count = ontology_mentions.get(ontology, 0)
                count += 1
                ontology_mentions[ontology] = count
ontology_mentions
```

```python
ontology_mentions_df = pd.DataFrame.from_dict(ontology_mentions, orient='index', columns=['count'])
```

```python
ontology_mentions_df.sort_values(by='count', ascending=False, inplace=True)
ontology_mentions_df.to_csv('data/ontology_mentions.csv')
```
